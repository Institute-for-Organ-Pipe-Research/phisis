## 1. Stała częstotliwość (Wykres 1)

- **Co jest pokazane:**  
  Klasyczna sinusoidalna fala o częstotliwości $f = 440 \,\mathrm{Hz}$ (ton a' z muzyki) generowana na cztery pełne okresy.

- **Matematyka:**  

  $$
  x(t) = \sin(2\pi f t), \quad f = 440\,\mathrm{Hz}, \quad t \in \left[0, \frac{4}{f}\right]
  $$

- **Cel:**  
  Pokazanie idealnej, stabilnej sinusoidy bez modulacji.

---

## 2. Liniowy chirp częstotliwości (Wykres 2)

- **Co jest pokazane:**  
  Sinusoida o częstotliwości zmieniającej się liniowo w czasie od $f_0 = 200\, \mathrm{Hz}$ do $f_1 = 1000\, \mathrm{Hz}$ w czasie $T = 0.015\, \mathrm{s}$.

- **Matematyka:**  

  $$f(t) = f_0 + k t, \quad k = \frac{f_1 - f_0}{T}$$

  $$
  \varphi(t) = 2\pi \int_0^t f(\tau)\,d\tau = 2\pi \left(f_0 t + \frac{1}{2} k t^2\right)
  $$

  $$
  x(t) = \sin(\varphi(t))
  $$

- **Cel:**  
  Pokazanie efektu modulacji częstotliwości (FM) w czasie rzeczywistym.

---

## 2b i 2c. Zoom na początek i koniec chirpa

- **Co jest pokazane:**  
  Szczegółowy widok fali w zakresie niskich i wysokich częstotliwości.

---

## 3. Skok częstotliwości ze sztucznym artefaktem (Wykres 3)

- **Co jest pokazane:**  
  Sygnał sinusoidalny, w którym częstotliwość skacze nagle z 330 Hz do 550 Hz, co powoduje niestabilność amplitudy.

- **Opis problemu:**  
  Klasyczne oscylatory cyfrowe nie kompensują zmiany fazy, przez co pojawiają się zakłócenia.

---

## 4. Skok częstotliwości bez artefaktu — opatentowane rozwiązanie (Wykres 4). Schemat na rys. 4

- **Co jest pokazane:**  
  Porównanie dwóch sygnałów:  
  - z artefaktem (klasyczny)  
  - bez artefaktu (z algorytmem patentowym)

- **Opis algorytmu:**

  Stabilizacja amplitudy podczas zmiany częstotliwości za pomocą wzmacniania zmiennej stanu i ograniczania amplitudy.

- **Algorytm:**

  $$
  \begin{aligned}
  \mathrm{VAR1}_{n} &= \mathrm{VAR1}_{n-1} - F^2 \cdot \mathrm{VAR2}_{n-1} \\
  \mathrm{VAR2}_{n} &= \mathrm{VAR2}_{n-1} \cdot (1 + \varepsilon) + \mathrm{VAR1}_{n}
  \end{aligned}
  $$

  $$
  \mathrm{VAR1} = \mathrm{clip}(\mathrm{VAR1}, -1, 1)
  $$

  $$
  F = 2 \sin\left(\frac{\pi f}{f_s}\right)
  $$

---

- **Opis:**

## 5. Porównanie widm globalnych

- **Opis:**  
  Widmo pełnego sygnału potwierdza stabilność wersji patentowej.

## Szczegóły implementacyjne oscylatora opatentowanego

Oscylator generuje sinusoidy poprzez dwie zmienne stanu i iteracyjne aktualizacje. Zapewnia stabilność amplitudy nawet przy skokowych zmianach częstotliwości.

- $f$ — bieżąca częstotliwość  
- $f_s$ — częstotliwość próbkowania  
- $\varepsilon \approx 0.001$ — czynnik wzmacniający energię

---

## Podsumowanie

- Klasyczne oscylatory mogą powodować rozstrajanie i zanikanie sygnału.
- Chirp pozwala analizować ciągłą modulację $f(t)$.
- Skoki częstotliwości wprowadzają artefakty.
- Opatentowany oscylator cyfrowy eliminuje te artefakty, stabilizując amplitudę.
- Widmo sygnału potwierdza skuteczność metody.

---

## Źródła i inspiracje

- [Patent US7442869B2](https://patents.google.com/patent/US7442869B2/en) – *Digital oscillator with amplitude stabilization*


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from scipy.fft import rfft, rfftfreq

fs = 44100  # częstotliwość próbkowania
plt.figure(figsize=(16, 28))
gs = GridSpec(6, 2, height_ratios=[1, 1, 1, 1, 1, 1])

# 1. STAŁA CZĘSTOTLIWOŚĆ
ax1 = plt.subplot(gs[0, :])
periods = 4
duration1 = periods / 440
t1 = np.linspace(0, duration1, int(fs * duration1))
signal_const = np.sin(2 * np.pi * 440 * t1)
ax1.plot(t1 * 1000, signal_const, 'b')
ax1.set_title("Wyk. 1:\nStabilna sinusoida 440 Hz (4 okresy)")
ax1.set_xlabel("Czas [ms]")
ax1.set_ylabel("Amplituda")
ax1.grid(True)
for i in range(periods + 1):
    ax1.axvline(i*1000/440, color='gray', linestyle=':', alpha=0.5)

# 2. LINIOWY CHIRP
duration2 = 0.015
t2 = np.linspace(0, duration2, int(fs * duration2))
f0, f1 = 200, 1000
k = (f1 - f0) / duration2
phi = 2 * np.pi * (f0 * t2 + 0.5 * k * t2**2)
chirp_signal = np.sin(phi)
ax2 = plt.subplot(gs[1, :])
ax2.plot(t2 * 1000, chirp_signal, 'g')
ax2.set_title("Wyk. 2:\nLiniowy wzrost częstotliwości (200 Hz → 1000 Hz)")
ax2.set_xlabel("Czas [ms]")
ax2.set_ylabel("Amplituda")
ax2.grid(True)
ax2.axvspan(0, 2, color='green', alpha=0.1, label='Zoom początku')
ax2.axvspan(13, 15, color='red', alpha=0.1, label='Zoom końca')
ax2.legend()

# 2b. ZOOM POCZĄTKU
ax2b = plt.subplot(gs[2, 0])
zoom_start = (t2 <= 0.002)
ax2b.plot(t2[zoom_start] * 1000, chirp_signal[zoom_start], 'g-o', markersize=3)
ax2b.set_title("Zoom:\nPoczątek chirpa (200 Hz)")
ax2b.set_xlabel("Czas [ms]")
ax2b.set_ylabel("Amplituda")
ax2b.grid(True)
for i in range(3):
    ax2b.axvline(i*5, color='green', linestyle=':', alpha=0.5)

# 2c. ZOOM KOŃCA
ax2c = plt.subplot(gs[2, 1])
zoom_end = (t2 >= 0.013)
ax2c.plot(t2[zoom_end] * 1000, chirp_signal[zoom_end], 'r-o', markersize=3)
ax2c.set_title("Zoom:\nKoniec chirpa (1000 Hz)")
ax2c.set_xlabel("Czas [ms]")
ax2c.set_ylabel("Amplituda")
ax2c.grid(True)
for i in range(3):
    ax2c.axvline(13 + i*1, color='red', linestyle=':', alpha=0.5)

# 3. SKOK Z ARTEFAKTEM
pre_time = 0.005
post_time = 0.005
duration3 = pre_time + post_time
t3 = np.linspace(0, duration3, int(fs * duration3))
f_jump = np.where(t3 < pre_time, 330, 550)
jump_signal = np.sin(2 * np.pi * f_jump * t3)
ax3 = plt.subplot(gs[3, :])
ax3.plot(t3 * 1000, jump_signal, 'r')
ax3.set_title("Wyk. 3:\nSkok częstotliwości Z artefaktem (330 Hz → 550 Hz)")
ax3.set_xlabel("Czas [ms]")
ax3.set_ylabel("Amplituda")
ax3.grid(True)
ax3.axvline(x=pre_time*1000, color='k', linestyle='--')
ax3.axvspan(4.5, 5.5, color='red', alpha=0.2)
ax3.text(5, 0.8, 'Zoom poniżej', ha='center', color='darkred')

# 4. SKOK BEZ ARTEFAKTU
def patent_oscillator(f_start, f_end, change_time, fs, duration):
    t_arr = np.arange(int(fs * duration)) / fs
    F_start = 2 * np.sin(np.pi * f_start / fs)
    F_end = 2 * np.sin(np.pi * f_end / fs)
    VAR1, VAR2 = 1.0, 0.0
    output = []
    for i in range(len(t_arr)):
        F = F_start if t_arr[i] < change_time else F_end
        VAR1 = VAR1 - F**2 * VAR2
        VAR2 = VAR2 * 1.001
        VAR2 = VAR1 + VAR2
        VAR1 = np.clip(VAR1, -1, 1)
        output.append(VAR1)
    return np.array(output)

patent_signal = patent_oscillator(330, 550, pre_time, fs, duration3)
zoom_mask = (t3 >= 0.0045) & (t3 <= 0.0055)

# 4b. ZOOM porównawczy
ax4a = plt.subplot(gs[4, 0])
ax4a.plot(t3[zoom_mask] * 1000, jump_signal[zoom_mask], 'ro-', label="Z artefaktem")
ax4a.plot(t3[zoom_mask] * 1000, patent_signal[zoom_mask], 'mo-', label="Bez artefaktu")
ax4a.set_title("Zoom:\nPorównanie sygnału przy skoku częstotliwości")
ax4a.set_xlabel("Czas [ms]")
ax4a.set_ylabel("Amplituda")
ax4a.grid(True)
ax4a.axvline(x=5.0, color='k', linestyle='--')
ax4a.legend()

# 4c. Widmo harmoniczne sygnału bez artefaktu
ax4b = plt.subplot(gs[4, 1])
yf = np.abs(rfft(jump_signal[zoom_mask]))
xf = rfftfreq(len(jump_signal[zoom_mask]), 1/fs)
yf_patent = np.abs(rfft(patent_signal[zoom_mask]))
xf_patent = rfftfreq(len(patent_signal[zoom_mask]), 1/fs)
ax4b.semilogy(xf, yf, 'r', label='Z artefaktem')
ax4b.semilogy(xf_patent, yf_patent, 'm', label='Bez artefaktu')
ax4b.set_title("Skład harmoniczny\nprzy skoku częstotliwości")
ax4b.set_xlabel("Częstotliwość [Hz]")
ax4b.set_ylabel("Amplituda (log)")
ax4b.grid(True)
ax4b.set_xlim(0, 5000)
ax4b.axvline(330, color='blue', linestyle=':', alpha=0.5)
ax4b.axvline(550, color='orange', linestyle=':', alpha=0.5)
ax4b.legend()

# 5. PORÓWNANIE WIDM (globalne)
ax5 = plt.subplot(gs[5, :])
ax5.semilogy(xf, yf, 'r', label='Z artefaktem')
ax5.semilogy(xf_patent, yf_patent, 'm', label='Bez artefaktu')
ax5.set_title("Wyk. 5:\nPorównanie składu harmonicznego (globalne)")
ax5.set_xlabel("Częstotliwość [Hz]")
ax5.set_ylabel("Amplituda (log)")
ax5.grid(True)
ax5.set_xlim(0, 5000)
ax5.axvline(330, color='blue', linestyle=':', alpha=0.5, label='330 Hz')
ax5.axvline(550, color='orange', linestyle=':', alpha=0.5, label='550 Hz')
ax5.legend()

plt.tight_layout()
plt.show()
```

## Sygnały 16 i 17

Sygnał 16 jest wyjściem z oscylatora cyfrowego 14. Z sygnału 16 tworzy się drugi sygnał 17 w bloku nr 15. Sygnał 17 to podwojona częstotliwość sygnału 16 określona wzorem po przekształceniach z tożsamości trygonometrycznych: sin(2ω0*n - π/2)

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametry sygnału
f0 = 440  # Częstotliwość podstawowa w Hz
fs = 44100  # Częstotliwość próbkowania
t = np.arange(0, 0.005, 1/fs)  # Czas od 0 do 5 ms (dla lepszej widoczności)

# Sygnał podstawowy: sin(2πf0t)
sine_wave = np.sin(2 * np.pi * f0 * t)

# Sygnał przekształcony: sin(2ω₀n - π/2)
# ω₀ = 2πf0/fs -> przekształcenie do jednostek dyskretnych
omega0 = 2 * np.pi * f0 / fs
n = np.arange(len(t))  # indeksy próbek
transformed_wave = np.sin(2 * omega0 * n - np.pi / 2)

# Tworzenie wykresów
fig, axs = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

# Wykres pierwszego sygnału
axs[0].plot(t * 1000, sine_wave, label='sin(2π·440·t)')
axs[0].set_title('Sygnał 16: Sinus 440 Hz')
axs[0].set_ylabel('Amplituda')
axs[0].legend()
axs[0].grid(True)

# Wykres drugiego sygnału
axs[1].plot(t * 1000, transformed_wave, label='sin(2ω₀·n - π/2)', color='orange')
axs[1].set_title('Sygnał 17: sin(2ω₀·n - π/2) = 880 Hz')
axs[1].set_xlabel('Czas [ms]')
axs[1].set_ylabel('Amplituda')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

Sygnały 16 i 17 są wzmacniane przez mnożniki: 18a - GAIN1 i 18b - GAIN2. Sygnały 18a, 18b przeczodzą przez hard limiter: 19a - CLIP1 I 19b - CLIP2.

Zakres wartości GAINx nie jest podana w patencie. AI sugeruje typowe wartości w praktyce w zakres 0.0 do 2.0, gdzie:

    - 0.0 = całkowite wyciszenie ścieżki,
    - 1.0 = nominalne wzmocnienie,
    - 1.0 = zwiększenie amplitudy (co zwiększa zniekształcenia przy clippingu).

Parametry te dobierane są empirycznie.

Funkcja CLIP określona jest wzorem:

$$
\text{CLIP}(x, \text{CLIP}_{\text{limit}}) = 
\begin{cases} 
-\text{CLIP}_{\text{limit}} & \text{dla } x < -\text{CLIP}_{\text{limit}} \\
x & \text{dla } |x| \leq \text{CLIP}_{\text{limit}} \\
+\text{CLIP}_{\text{limit}} & \text{dla } x > +\text{CLIP}_{\text{limit}}
\end{cases}
$$

Dla sygnałów w Fig. 3 patentu:

$$\text{Ścieżka podstawowa (19a):}\quad y_1[n] = \text{CLIP}\left( \text{GAIN}_1 \cdot \sin(\omega_0 n), \text{CLIP}_1 \right)$$

$$\text{Ścieżka podwójnej częst. (19b):}\quad y_2[n] = \text{CLIP}\left( \text{GAIN}_2 \cdot \sin(2\omega_0 n - \frac{\pi}{2}), \text{CLIP}_2 \right)$$

Prawdopodobny przedział wartości dla CLIP1/CLIP2 ∈ (0.1, 1.0). Domyślnie prawdopodobnie ~0.7.

W związku z tym, że GAINx i CLIPx są ze sobą powiązane ważne są stosunki tych parametrów. Colin Pykett w swojej analizie zauważa:

> *"Dokładne wartości GAIN1/GAIN2 i CLIP1/CLIP2 są parametrami brzmienia dostosowanymi empirycznie, aby naśladować określone typy głosów organowych"*.

Podaje też przykłady:

    - Dla Principal: CLIP1 ≈ 0.8, CLIP2 ≈ 0.6,
    - Dla Fletu krytego: CLIP1 ≈ 0.9, CLIP2 ≈ 0.3.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parametry
f0 = 440
fs = 44100
t = np.arange(0, 0.01, 1/fs)  # 10 ms
n = np.arange(len(t))

omega0 = 2 * np.pi * f0 / fs

# Parametry eksperymentu (możesz zmieniać)
GAIN1 = 1.0
GAIN2 = 0.5
CLIP1 = 0.8
CLIP2 = 0.3

# Funkcja przycinająca (clipping)
def clip(x, threshold):
    return np.clip(x, -threshold, threshold)

# Sygnały
s1 = clip(GAIN1 * np.sin(omega0 * n), CLIP1)
s2 = clip(GAIN2 * np.sin(2 * omega0 * n - np.pi / 2), CLIP2)

# Sygnał sumaryczny
s_total = s1 + s2

# Wykresy czasowe
plt.figure(figsize=(12, 6))
plt.subplot(3, 1, 1)
plt.plot(t * 1000, s1, label='output_19a (nieparzyste)', color='blue')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t * 1000, s2, label='output_19b (parzyste)', color='orange')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t * 1000, s_total, label='suma', color='green')
plt.legend()
plt.xlabel("Czas [ms]")
plt.grid(True)

plt.tight_layout()
plt.show()

# Analiza FFT
def plot_fft(signal, fs, title):
    N = len(signal)
    freq = np.fft.rfftfreq(N, 1/fs)
    fft_mag = np.abs(np.fft.rfft(signal)) / N
    plt.plot(freq, 20 * np.log10(fft_mag + 1e-10), label=title)

plt.figure(figsize=(10, 4))
plot_fft(s1, fs, 'Nieparzyste (output_19a)')
plot_fft(s2, fs, 'Parzyste (output_19b)')
plot_fft(s_total, fs, 'Suma')
plt.title("Widma sygnałów")
plt.xlabel("Częstotliwość [Hz]")
plt.ylabel("Amplituda [dB]")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 44100
f0 = 440
t = np.arange(0, 0.05, 1/fs)  # dłuższy czas = lepsza rozdzielczość widmowa
n = np.arange(len(t))
omega0 = 2 * np.pi * f0 / fs

def clip(x, limit):
    return np.clip(x, -limit, limit)

# Lista wariantów: (nazwa, G1, C1, G2, C2)
variants = [
    ("Principal",         1.0, 0.8, 0.8, 0.6),
    ("Flet kryty",        1.0, 0.9, 0.7, 0.3),
    ("Eksperymentalny",   1.5, 0.5, 1.2, 0.2)
]

# Funkcja do ekstrakcji amplitud harmonicznych
def harmonic_amplitudes(signal, fs, f0, N=10):
    fft = np.fft.rfft(signal * np.hanning(len(signal)))  # okno Hanninga
    mag = np.abs(fft) / len(signal)
    freqs = np.fft.rfftfreq(len(signal), 1/fs)
    amps = []
    for k in range(1, N+1):
        idx = np.argmin(np.abs(freqs - k*f0))
        amps.append(mag[idx])
    return amps

# Rysowanie
fig, axs = plt.subplots(len(variants), 3, figsize=(16, 9))
fig.suptitle("Porównanie: przebieg + widmo + rozkład harmonicznych", fontsize=16)

for i, (name, G1, C1, G2, C2) in enumerate(variants):
    y1 = clip(G1 * np.sin(omega0 * n), C1)
    y2 = clip(G2 * np.sin(2 * omega0 * n - np.pi/2), C2)
    y_sum = y1 + y2

    # === 1. Przebieg czasowy ===
    axs[i][0].plot(t * 1000, y_sum, color='green')
    axs[i][0].set_title(f"{name} – sygnał czasowy")
    axs[i][0].set_ylabel("Amplituda")
    axs[i][0].grid(True)

    # === 2. Widmo (dB) ===
    Nfft = len(y_sum)
    freqs = np.fft.rfftfreq(Nfft, 1/fs)
    fft_mag = np.abs(np.fft.rfft(y_sum * np.hanning(len(y_sum)))) / Nfft
    axs[i][1].plot(freqs, 20 * np.log10(fft_mag + 1e-12), color='purple')
    axs[i][1].set_xlim(0, 5000)
    axs[i][1].set_title("Widmo sygnału [dB]")
    axs[i][1].set_ylabel("Amplituda [dB]")
    axs[i][1].grid(True)

    # === 3. Rozkład harmonicznych ===
    amps = harmonic_amplitudes(y_sum, fs, f0, N=10)
    harmonics = [f"{k}f₀" for k in range(1, 11)]
    axs[i][2].bar(harmonics, amps, color='orange')
    axs[i][2].set_title("Rozkład 10 pierwszych harmonicznych")
    axs[i][2].set_ylabel("Amplituda")
    axs[i][2].set_ylim(0, max(amps)*1.1)
    axs[i][2].grid(True)

    # Podpis osi czasu tylko na dole
    if i == len(variants) - 1:
        axs[i][0].set_xlabel("Czas [ms]")
        axs[i][1].set_xlabel("Częstotliwość [Hz]")
        axs[i][2].set_xlabel("Harmoniczna")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

## Interaktywne narzędzie

**Uwaga!** Dźwięk odtwarza się automatycznie

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, IntSlider, VBox, HBox, Output
from IPython.display import display, clear_output, Audio

fs = 44100
f0 = 440

def clip(x, limit):
    return np.clip(x, -limit, limit)

def apply_ar_envelope(signal, fs, attack_time, release_time):
    N = len(signal)
    a_samples = int(attack_time * fs)
    r_samples = int(release_time * fs)
    sustain_samples = N - a_samples - r_samples
    if sustain_samples < 0:
        raise ValueError("Czas ataku i wygaszenia za duży względem długości sygnału")
    attack = np.linspace(0, 1, a_samples, endpoint=False)
    sustain = np.ones(sustain_samples)
    release = np.linspace(1, 0, r_samples, endpoint=True)
    envelope = np.concatenate([attack, sustain, release])
    return signal * envelope, (a_samples, a_samples + sustain_samples)

def harmonic_amplitudes(signal, fs, f0, N=16):
    fft = np.fft.rfft(signal * np.hanning(len(signal)))
    mag = np.abs(fft) / len(signal)
    freqs = np.fft.rfftfreq(len(signal), 1/fs)
    amps = []
    for k in range(1, N+1):
        idx = np.argmin(np.abs(freqs - k*f0))
        amps.append(mag[idx])
    return amps

out = Output()

def plot_and_prepare_audio(GAIN1, CLIP1, GAIN2, CLIP2, duration, harmonic_count, attack, release):
    t = np.arange(0, duration, 1/fs)
    n = np.arange(len(t))
    omega0 = 2 * np.pi * f0 / fs

    y1 = clip(GAIN1 * np.sin(omega0 * n), CLIP1)
    y2 = clip(GAIN2 * np.sin(2 * omega0 * n - np.pi/2), CLIP2)
    y_sum = y1 + y2
    y_env, (start_idx, end_idx) = apply_ar_envelope(y_sum, fs, attack, release)

    # Faza sustain - pełna do analizy
    y_sustain = y_env[start_idx:end_idx]
    t_sustain = t[start_idx:end_idx]

    # Krótki fragment (5 okresów) do wykresu sygnału
    period_samples = int(fs / f0)
    fragment_samples = 5 * period_samples
    if fragment_samples > len(y_sustain):
        fragment_samples = len(y_sustain)
    y_fragment = y_sustain[:fragment_samples]
    t_fragment = t_sustain[:fragment_samples]

    Nfft = len(y_sustain)
    freqs = np.fft.rfftfreq(Nfft, 1/fs)
    fft_mag = np.abs(np.fft.rfft(y_sustain * np.hanning(Nfft))) / Nfft
    amps = harmonic_amplitudes(y_sustain, fs, f0, N=harmonic_count)
    harmonics = [f"H{k}" for k in range(1, harmonic_count + 1)]

    with out:
        clear_output(wait=True)
        plt.figure(figsize=(16, 5))

        plt.subplot(1, 3, 1)
        plt.plot(t_fragment * 1000, y_fragment, color='green')
        plt.title("Sygnał czasowy (pierwsze 5 okresów fazy sustain)")
        plt.xlabel("Czas [ms]")
        plt.ylabel("Amplituda")
        plt.grid(True)
        plt.xlim(t_fragment[0]*1000, t_fragment[-1]*1000)
        ymin, ymax = np.min(y_fragment), np.max(y_fragment)
        y_margin = (ymax - ymin) * 0.1
        plt.ylim(ymin - y_margin, ymax + y_margin)

        plt.subplot(1, 3, 2)
        plt.plot(freqs, 20 * np.log10(fft_mag + 1e-12), color='purple')
        plt.xlim(0, 5000)
        plt.title("Widmo [dB] (faza sustain)")
        plt.xlabel("Częstotliwość [Hz]")
        plt.grid(True)

        plt.subplot(1, 3, 3)
        bars = plt.bar(harmonics, amps, color='orange')
        plt.title(f"Rozkład {harmonic_count} harmonicznych")
        plt.ylabel("Amplituda")
        plt.xticks(rotation=45)
        plt.grid(True)
        for bar, label in zip(bars, harmonics):
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, height + 0.001, label,
                     ha='center', va='bottom', fontsize=8)

        plt.tight_layout()
        plt.show()

    # Przygotowanie audio - normalizacja i konwersja do int16
    audio_clean = np.nan_to_num(y_env, nan=0.0, posinf=0.0, neginf=0.0)
    max_val = np.max(np.abs(audio_clean))
    if max_val > 0:
        audio_norm = audio_clean / max_val
    else:
        audio_norm = audio_clean
    audio_int16 = (audio_norm * 32767).astype(np.int16)

    return audio_int16

params = {
    "GAIN1": 1.0, "CLIP1": 0.8, "GAIN2": 0.8, "CLIP2": 0.6,
    "duration": 0.5, "harmonic_count": 16, "attack": 0.01, "release": 0.01
}

GAIN1_slider = FloatSlider(value=params["GAIN1"], min=0, max=2, step=0.1, description='GAIN1')
CLIP1_slider = FloatSlider(value=params["CLIP1"], min=0.1, max=1, step=0.05, description='CLIP1')
GAIN2_slider = FloatSlider(value=params["GAIN2"], min=0, max=2, step=0.1, description='GAIN2')
CLIP2_slider = FloatSlider(value=params["CLIP2"], min=0.1, max=1, step=0.05, description='CLIP2')
duration_slider = FloatSlider(value=params["duration"], min=0.1, max=5, step=0.1, description='Czas [s]')
harmonic_slider = IntSlider(value=params["harmonic_count"], min=4, max=64, step=1, description='Harmonics')
attack_slider = FloatSlider(value=params["attack"], min=0.005, max=0.2, step=0.005, description='Attack [s]')
release_slider = FloatSlider(value=params["release"], min=0.005, max=0.2, step=0.005, description='Release [s]')

audio_data = None

def on_change(change):
    global audio_data
    audio_data = plot_and_prepare_audio(
        GAIN1_slider.value, CLIP1_slider.value,
        GAIN2_slider.value, CLIP2_slider.value,
        duration_slider.value, harmonic_slider.value,
        attack_slider.value, release_slider.value
    )
    with out:
        display(Audio(data=audio_data, rate=fs, autoplay=True))

for w in [GAIN1_slider, CLIP1_slider, GAIN2_slider, CLIP2_slider,
          duration_slider, harmonic_slider, attack_slider, release_slider]:
    w.observe(on_change, names='value')

on_change(None)  # startowa inicjalizacja

ui = VBox([
    HBox([GAIN1_slider, CLIP1_slider]),
    HBox([GAIN2_slider, CLIP2_slider]),
    HBox([duration_slider, harmonic_slider]),
    HBox([attack_slider, release_slider]),
    out
])

display(ui)
```
